```python
from sklearn import cross_validation as cv
train_data, test_data = cv.train_test_split(trainDF, test_size=0.25)
```

```python
train_data.describe()
```

```python
test_data.describe()
```

```python
train_data_matrix = np.zeros((user_row_size, user_row_size))
```

```python
train_data_matrix.size
```

```python
for line in train_data.itertuples():
    train_data_matrix[line[1]-1, line[2]-1] = line[3]

test_data_matrix = np.zeros((user_row_size, user_row_size))
for line in test_data.itertuples():
    test_data_matrix[line[1]-1, line[2]-1] = line[3]
```

```python
train_data_matrix
```

```python
test_data_matrix
```

```python
from sklearn.metrics.pairwise import pairwise_distances
user_similarity = pairwise_distances(train_data_matrix, metric='cosine')
#item_similarity = pairwise_distances(train_data_matrix.T, metric='cosine')
```

```python
user_similarity
```

```python
def predict(ratings, similarity):
    mean_user_rating = ratings.mean(axis=1)
    rating_diff = (ratings - mean_user_rating[:, np.newaxis])
    pred = mean_user_rating[:, np.newaxis] + similarity.dot(rating_diff) / np.array([np.abs(similarity).sum(axis=1)]).T
    return pred
```

```python
user_prediction = predict(train_data_matrix, user_similarity)
```

```python
user_prediction
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
def rmse(prediction, ground_truth):
    prediction = prediction[ground_truth.nonzero()].flatten()
    ground_truth = ground_truth[ground_truth.nonzero()].flatten()
    return sqrt(mean_squared_error(prediction, ground_truth))
```

```python
print ('User-based CF RMSE: ' + str(rmse(user_prediction, test_data_matrix)))
```
